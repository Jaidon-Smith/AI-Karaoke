````python
!mkdir opusdownloads
!mkdir textdownloads
!mkdir opus
!mkdir texts
files = os.listdir('manifests')
for file in files:
  file = file.split('.')[0]
  

  url = "https://azureopendatastorage.blob.core.windows.net/openstt/ru_open_stt_opus_unpacked/"

  duplicate_check = {}

  count = 1
  with open('manifests/' + file + '.csv', 'r') as f:
    opus_folder = 'opus/' + file
    txt_folder = 'texts/' + file

    !mkdir $opus_folder
    !mkdir $txt_folder
    print('reading file', file)
    lines = f.readlines()
    for line in lines:
      line = line.replace('\n', '')
      line = line.split(',')
      opus = line[0]
      opus_filename = ''.join(opus.split('/')[-3:])
      txt = line[1]
      txt_filename = ''.join(txt.split('/')[-3:])
      if opus_filename in duplicate_check:
        #print('Deleting duplicate', opus_filename)
        continue
      duplicate_check[opus_filename] = True
      #print(opus_filename, txt_filename)
      #download(url + opus, out='/content/opus/' + opus_filename)
      #download(url + txt, out='/content/txt/' + txt_filename)
      
      #print((curr_time-original_time)/count)
      count += 1
      
      with open('opusdownloads/' + file + '.txt', 'a+') as d:
        d.write(url + opus + '\n')
        d.write(' dir=' + opus_folder + '\n out=' + opus_filename + '\n')
      with open('textdownloads/file' + file + '.txt', 'a+') as d:
        d.write(url + txt + '\n')
        d.write(' dir=' + txt_folder + '\n out=' + txt_filename + '\n')
        #d.write(' dir=text\n out=' + txt_filename + '\n')
      if count == 100:
        pass
    
````
